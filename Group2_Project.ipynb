Missing Values

```python
# Count missing values per column
missing_counts = CMS_data.isnull().sum()
missing_percent = (missing_counts / len(CMS_data)) * 100
# Show only columns with missing data
missing_summary = pd.DataFrame({
'Missing Values': missing_counts,
'Percent Missing': missing_percent
})
missing_summary = missing_summary[missing_summary['Missing Values'] > 0]
missing_summary.sort_values(by='Percent Missing', ascending=False)
```

CMS Data Preprocessing

```python
# Show missing values again (recap)
CMS_data.isnull().sum().sort_values(ascending=False).head(10)
```

```python
# Convert Rndrng_Prvdr_Geo_Cd (FIPS code) to string so it can be treated as a categorical variable
CMS_data['Rndrng_Prvdr_Geo_Cd'] = CMS_data['Rndrng_Prvdr_Geo_Cd'].astype(str)

# Fill missing FIPS codes with 'Unknown' to preserve rows and treat them as a separate categories
CMS_data['Rndrng_Prvdr_Geo_Cd'].fillna('Unknown', inplace=True)

# Fill missing state names (Rndrng_Prvdr_Geo_Desc) with 'Unknown' to avoid dropping rows
CMS_data['Rndrng_Prvdr_Geo_Desc'].fillna('Unknown', inplace=True)
```

```python
# Cap Avg_Sbmtd_Chrg at 99th percentile to reduce impact of extreme outliers
cap_value = CMS_data['Avg_Sbmtd_Chrg'].quantile(0.99)
CMS_data['Avg_Sbmtd_Chrg_Capped'] = CMS_data['Avg_Sbmtd_Chrg'].apply(lambda x: min(x, cap_value))
```

##SVM model

```python
X = CMS_data['Avg_Sbmtd_Chrg_Capped']
y = CMS_data.target

# Split data into training and testing sets
X_train, X_test, y_train, y_test = skm(X, y, test_size = 0.2, random_state = 0)
```

```python
#SVM classifier
CMS_svm = SVC(kernel='linear', C=1)

#Train the model using the training data
CMS_svm.fit(X_train, y_train)
```

```python
# Predicting the labels for the test data
CMS_ypred = CMS_svm.predict(X_test)
```

```python
#Calculate the accuracy
CMS_acc = RocCurveDisplay(y_test, CMS_ypred)
print("Accuracy:", CMS_acc)
```

```python
# TAN NGUYEN - XGBoost

# Loading Data
import os

# Root folder where all subfolders live
root = "./data"

# Names of subfolders
subfolders = ['CMS', 'Kaggle', 'Synthea']

# Master dictionary to store datasets
all_data = {}

# Loop through each subfolder
for subfolder in subfolders:
    folder_path = os.path.join(root, subfolder)
    dataset_dict = {}
    for filename in os.listdir(folder_path):
        if filename.endswith(".csv"):
            file_path = os.path.join(folder_path, filename)
            df_name = filename.replace(".csv", "")
            dataset_dict[df_name] = pd.read_csv(file_path)
            print(f"✅ Loaded {df_name} from {subfolder}")
    all_data[subfolder] = dataset_dict

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from xgboost import XGBClassifier  # Use RandomForestClassifier if XGBoost fails

# -----------------------------
# Load datasets
# -----------------------------

# Load the main training dataset (has fraud labels)
kaggle_train_df = all_data['Kaggle']['Train-1542865627584']
df_target = all_data['Kaggle']['Train-1542865627584']
df_inpatient = all_data['Kaggle']['Train_Inpatientdata-1542865627584']
df_outpatient = all_data['Kaggle']['Train_Outpatientdata-1542865627584']
df_beneficiary = all_data['Kaggle']["Train_Beneficiarydata-1542865627584"]
```

```python
## Tan Nguyen

## Training Using only data from inpatient, and outpatient

# -----------------------------
# Feature engineering function
# -----------------------------
def preprocess_claim_data(df, provider_col="Provider", is_inpatient=True):
    date_start = 'AdmissionDt' if is_inpatient else 'ClaimStartDt'
    date_end = 'DischargeDt' if is_inpatient else 'ClaimEndDt'

    df[date_start] = pd.to_datetime(df[date_start], errors='coerce')
    df[date_end] = pd.to_datetime(df[date_end], errors='coerce')
    df['ClaimDuration'] = (df[date_end] - df[date_start]).dt.days.fillna(0)

    # Aggregate features
    agg = df.groupby(provider_col).agg({
        'ClaimID': 'count',
        'InscClaimAmtReimbursed': 'sum',
        'ClaimDuration': ['mean', 'sum']
    })
    agg.columns = [f"{'In' if is_inpatient else 'Out'}_{col[0]}_{col[1]}" for col in agg.columns]
    agg.reset_index(inplace=True)
    return agg

# -----------------------------
# Generate features
# -----------------------------
inpatient_feat = preprocess_claim_data(df_inpatient, is_inpatient=True)
outpatient_feat = preprocess_claim_data(df_outpatient, is_inpatient=False)

# Merge features
features = pd.merge(inpatient_feat, outpatient_feat, on="Provider", how="outer").fillna(0)

# Merge with target
df = pd.merge(df_target, features, on="Provider", how="left").fillna(0)
df['fraud_label'] = (df['PotentialFraud'] == 'Yes').astype(int)

# -----------------------------
# Train/Test split
# -----------------------------
X = df.drop(columns=["Provider", "PotentialFraud", "fraud_label"])
y = df["fraud_label"]
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=42, test_size=0.2)

# -----------------------------
# Model training
# -----------------------------
try:
    model = XGBClassifier(n_estimators=100, learning_rate=0.1, max_depth=4, use_label_encoder=False, eval_metric='logloss')
    model.fit(X_train, y_train)
except:
    from sklearn.ensemble import RandomForestClassifier
    model = RandomForestClassifier()
    model.fit(X_train, y_train)

# -----------------------------
# Model evaluation
# -----------------------------
y_pred = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
## Tan Nguyen
## Training using inpatient, outpatient, beneficiary

# -----------------------------
# Feature engineering: Claims
# -----------------------------
def preprocess_claim_data(df, is_inpatient=True):
    start = 'AdmissionDt' if is_inpatient else 'ClaimStartDt'
    end = 'DischargeDt' if is_inpatient else 'ClaimEndDt'
    df[start] = pd.to_datetime(df[start], errors='coerce')
    df[end] = pd.to_datetime(df[end], errors='coerce')
    df['ClaimDuration'] = (df[end] - df[start]).dt.days.fillna(0)
    agg = df.groupby('Provider').agg({
        'ClaimID': 'count',
        'InscClaimAmtReimbursed': 'sum',
        'ClaimDuration': ['mean', 'sum']
    })
    agg.columns = [f"{'In' if is_inpatient else 'Out'}_{c}_{f}" for c, f in agg.columns]
    agg.reset_index(inplace=True)
    return agg

in_feat = preprocess_claim_data(df_inpatient, is_inpatient=True)
out_feat = preprocess_claim_data(df_outpatient, is_inpatient=False)
features = pd.merge(in_feat, out_feat, on='Provider', how='outer').fillna(0)

# -----------------------------
# Feature engineering: Beneficiary
# -----------------------------
pairs = pd.concat([
    df_inpatient[['Provider', 'BeneID']],
    df_outpatient[['Provider', 'BeneID']]
]).drop_duplicates().groupby(['Provider', 'BeneID']).head(1)

bene_cols = ['BeneID', 'DOB', 'Gender', 'Race', 'RenalDiseaseIndicator',
             'NoOfMonths_PartACov', 'NoOfMonths_PartBCov'] + \
            [col for col in df_beneficiary.columns if col.startswith('ChronicCond_')]

df_bene_small = df_beneficiary[bene_cols]
bene_merged = pd.merge(pairs, df_bene_small, on='BeneID', how='left')

bene_merged['DOB'] = pd.to_datetime(bene_merged['DOB'], errors='coerce')
bene_merged['Age'] = 2009 - bene_merged['DOB'].dt.year
bene_merged['RenalDisease'] = bene_merged['RenalDiseaseIndicator'].map({'Y': 1, 'N': 0})

agg_dict = {
    'Age': 'mean',
    'Gender': 'mean',
    'Race': 'mean',
    'RenalDisease': 'mean',
    'NoOfMonths_PartACov': 'mean',
    'NoOfMonths_PartBCov': 'mean'
}
for col in bene_merged.columns:
    if col.startswith("ChronicCond_"):
        agg_dict[col] = 'mean'

provider_bene = bene_merged.groupby('Provider').agg(agg_dict).reset_index()
provider_bene.columns = ['Provider'] + [f"Bene_{col}" for col in provider_bene.columns if col != 'Provider']

# -----------------------------
# Merge everything
# -----------------------------
all_features = pd.merge(features, provider_bene, on='Provider', how='left').fillna(0)
df = pd.merge(df_target, all_features, on='Provider', how='left').fillna(0)
df['fraud_label'] = (df['PotentialFraud'] == 'Yes').astype(int)

X = df.drop(columns=['Provider', 'PotentialFraud', 'fraud_label'])
y = df['fraud_label']
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=42)

# -----------------------------
# Train & Evaluate Model
# -----------------------------
model = XGBClassifier(n_estimators=100, learning_rate=0.1, max_depth=4, use_label_encoder=False, eval_metric='logloss')
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
print("✅ Accuracy:", accuracy_score(y_test, y_pred))
print("✅ Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("✅ Classification Report:\n", classification_report(y_test, y_pred))
```
